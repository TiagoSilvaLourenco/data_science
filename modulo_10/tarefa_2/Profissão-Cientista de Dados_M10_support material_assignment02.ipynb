1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
````

````python
df = pd.read_csv('../previsao_de_renda.csv')

df.head()
````

> ## Gráfico posse de veículo X renda <span id='posse_de_veiculo'></span>

<div style='text-align:right; width:100%; '>
    <a style="color:bluemarine; background-color:whitesmoke; padding:5px 10px; border-radius:7px; font-size:1.6em" href='#insight'>Insight</a>
</div>

````python
ax = sns.pointplot(data=df, x="posse_de_veiculo", y="renda",
                  errorbar=("pi", 100), capsize=.4, join=False, color=".5",
                  )
````

> #### O gráfico acima mostra que quem tem a posse de veículo possui renda superior a quem não tem carro. 

````python
ax = sns.pointplot(x="posse_de_imovel", y="renda", data=df,
                  errorbar=("pi", 100), capsize=.4, join=False, color=".5"
                  )
````

> #### O gráfico acima mostra que quem possui imóvel próprio tem uma renda superior.

````python
ax = sns.pointplot(data=df, x="posse_de_veiculo", y="renda", hue='posse_de_imovel', dodge=True,
                  errorbar=("pi", 100), capsize=.4, join=False, color=".8"
                  )
````

## Insight <span id='insight'></a>

> Analisando os gráficos acima, ambas variáveis são boas preditoras de renda, mas a variável **<span><a style='color:red' href='#posse_de_veiculo'>posse_de_veiculo</a></span>** possui uma discrepância estatística maior, onde quem não possui veículo tem uma renda que fica entre **4.300** e **4.500** e quem possui veículo tem uma renda que fica entre **4.900** e **5.100**
