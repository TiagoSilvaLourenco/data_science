````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
````

````python
df = pd.read_csv('../previsao_de_renda.csv')

df.head()
````

1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
ax = sns.countplot(x='data_ref',  hue = 'tipo_residencia', data=df)

tick_labs = df['data_ref'].map(lambda ts: ts).unique()

# tick_labs.tolist()

ticks = ax.set_xticks(list(range(df['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

````python
ax = sns.countplot(x='data_ref',  hue = 'estado_civil', data=df)

tick_labs = df['data_ref'].map(lambda ts: ts).unique()

# tick_labs.tolist()

ticks = ax.set_xticks(list(range(df['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

````python
ax = sns.countplot(x='data_ref',  hue = 'educacao', data=df)

tick_labs = df['data_ref'].map(lambda ts: ts).unique()

# tick_labs.tolist()

ticks = ax.set_xticks(list(range(df['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

> ### Há uma alteração nessa variavel, a partir de 01/09/2015 onde que quem tem educação secundária o indice cai e superior tem uma subida no mesmo período, os demais indices não se alteram sua trajetoria no tempo.
